# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```

```python

# Check for missing values and data types

print("Missing values in each column:\n")
print(df.isnull().sum(), "\n")

# Check data types
print("Data types of each column:\n")
print(df.dtypes)
```

```python

# Summary statistics of numerical columns

print("Summary statistics of numerical columns:\n")
print(df.describe(), "\n")
```

```python
# Summary statistics of categorical columns

print("Summary statistics of categorical columns:\n")
print(df.describe(include=['object']))
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable

# Load dataset
df = pd.read_csv("bankmarketing.csv")

# Display the column names to identify the target variable
print("Columns in dataset:\n", df.columns, "\n")

# Assuming the target column is named 'y' (typical in bank marketing dataset)
target_col = 'y'

# Check unique values in the target column
print("Unique values in target column:\n", df[target_col].value_counts(), "\n")

# Plot distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=target_col, hue=target_col, data=df, palette="Set2")
plt.title("Distribution of Target Variable")
plt.xlabel("Target (Subscription to Term Deposit)")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Correlation matrix for numerical features

# Load dataset
df = pd.read_csv("bankmarketing.csv")

# Select only numerical columns
num_df = df.select_dtypes(include=['int64', 'float64'])

# Display correlation matrix
print("Correlation matrix:\n")
corr_matrix = num_df.corr()
print(corr_matrix, "\n")

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix for Numerical Features")
plt.show()
```
